# Exponential Attenuation Law

X-ray attenuation follows the **Beer-Lambert Law**:

$$
I = I_0 \cdot e^{-\mu x}
$$

where:
- $I$ is the transmitted intensity after passing through material thickness $x$,
- $I_0$ is the initial intensity of the X-ray beam,
- $\mu$ is the **linear attenuation coefficient** (dependent on material and energy),
- $x$ is the thickness of the material.

The attenuation coefficient $\mu$ determines how strongly a material absorbs X-rays.
